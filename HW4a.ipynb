```python
#Problem 2: 1D case

import numpy as np
import matplotlib.pyplot as plt

Nloops = 10000
position_x = np.zeros(Nloops)

N_array = np.array([2, 4, 8, 10, 16, 20, 30, 40, 50, 75, 100])
rms = np.zeros(N_array.size)    
var = np.zeros(N_array.size) 

for count in range(N_array.size):
    N = N_array[count]
    for i in range(Nloops):
        x = 0.
        x1 = 0.
        x2 = 0.
        step = 0.
        for j in range(N):
            step_size = np.random.exponential(scale=1.)
            a = np.random.randint(0,1)
            if(a==0):
                x = x - step_size
                x2 = x2 + 1
            else:
                x = x + step_size 
                x1 = x1 + 1

            step += step_size 
            
        position_x[i] = x
        step_avg = step/N
        
    var[count] = np.var(position_x)
    rms[count] = np.sqrt(np.var(position_x))

plt.scatter(N_array, rms)   
plt.plot(N_array, np.sqrt(N_array)*np.sqrt(step_avg), "r")   #the line show f = 1.5*sqrt(N)
plt.xlabel("Number of steps N")
plt.ylabel("RMS")
plt.show()

plt.scatter(N_array, var)
plt.xlabel("Number of steps")
plt.ylabel("Variance")
plt.plot(N_array, N_array*step_avg, "r")
coef = np.polyfit(N_array, var, 1)
print(coef)
plt.show()
```
